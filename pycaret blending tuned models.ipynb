```python
import numpy as np 
import pandas as pd 
from pycaret.regression import *    
```

```python
data = pd.read_csv("3_features_predict.csv")
```

```python
reg = setup(data = data, target = '머리너비', session_id=123)
```

튜닝은 CPU에서 시행하면 시간이 오래 걸리기때문에, 훈련속도가 빠른 아래 6가지 모델을 활용하기로 결정했다. 
1. Catboost
2. Light GBM
3. Linear Regression 
4. Ridge 
5. Least Angle Regression 
6. Bayesian Ridge 

## Tune Catboost

```python
catboost = create_model('catboost')
```

```python
tuned_catboost = tune_model(catboost, n_iter = 100) # random grid search  
```

## Tune LGBM 

```python
lgbm = create_model('lightgbm')
```

```python
tuned_lgbm = tune_model(lgbm, n_iter = 100) 
```

## Tune linear regression

```python
lr = create_model('lr') 
```

```python
tuned_lr = tune_model(lr, n_iter = 100) 
```

## Tune Ridge

```python
ridge = create_model('ridge') 
```

```python
tuned_ridge = tune_model(ridge, n_iter = 100)
```

## Least Angle Regression

```python
lar = create_model('lar')
```

```python
tuned_lar = tune_model(lar, n_iter = 100)
```

## Bayesian Ridge

```python
br = create_model('br')
```

```python
tuned_br = tune_model(br, n_iter = 100)
```

```python
models = []
models.append(tuned_lgbm)
models.append(tuned_catboost) 
models.append(tuned_lr) 
models.append(tuned_ridge)
models.append(tuned_lar)
models.append(tuned_br)
```

```python
blended = blend_models(estimator_list = models, fold = 5)
```

```python
predict_holdout = predict_model(blended)
```
